```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, roc_auc_score, precision_recall_curve, precision_score, accuracy_score
from sklearn.metrics import recall_score, classification_report, f1_score, roc_curve, auc
```

```python
data = pd.read_csv('titanic_clean.csv')
df = data.copy()
```

```python
df.shape
```

```python
df.head()
```

```python
X = df.drop('Survived', axis=1)
Y = df['Survived']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size=0.3)
print(xtrain.shape, ytrain.shape)
print(xtest.shape, ytest.shape)
```

```python
from sklearn import preprocessing
#creating labelEncoder
le = preprocessing.LabelEncoder()
```

```python
#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
model = GaussianNB()

# Train the model using the training sets
model.fit(xtrain,ytrain)

```

```python
xtest
```

```python
#Predict Output
predictedLabel = model.predict(xtest) 
print ("Predicted Value:", predictedLabel)
```

```python
from sklearn import metrics  
  
# Model Accuracy, how often is the classifier correct?  
print("Accuracy:",metrics.accuracy_score(ytest, predictedLabel)) 
```
